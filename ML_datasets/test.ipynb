```python
image = pd.read_csv('image_category.csv')
```

```python
image.columns
```

```python
print(image['Image Urls'].iloc[100])
# the seperator is |
```

```python
top_categories = (
    image['Category']
    .value_counts()
    .head(10)
    .index
)

image_df = image[image['Category'].isin(top_categories)]
image_df.head()
```

```python
image_df['Category'].unique()
```

```python
MAX_IMAGES_PER_CATEGORY = 100

image_df = (
    image_df
    .groupby('Category')
    .apply(lambda x: x.sample(
        n=min(len(x), MAX_IMAGES_PER_CATEGORY),
        random_state=42
    ))
    .reset_index(drop=True)
)
```

```python
import re

def extract_high_res_image(urls):
    if pd.isna(urls):
        return None

    first_url = urls.split('|')[0]

    # Remove Amazon size modifier like ._SS40_
    high_res_url = re.sub(r'\._.*?_\.', '.', first_url)

    return high_res_url



image_df['Primary_Image'] = image_df['Image Urls'].apply(extract_high_res_image)
image_df = image_df.dropna(subset=['Primary_Image'])
```

```python
image_df['Primary_Image'].iloc[0]
```

```python
import os
import requests

def download_images(df, category_col, url_col):
    for _, row in df.iterrows():
        category = row[category_col]
        url = row[url_col]

        folder = f"images/{category}"
        os.makedirs(folder, exist_ok=True)

        try:
            img_data = requests.get(url, timeout=10).content
            filename = os.path.join(folder, url.split("/")[-1])
            with open(filename, "wb") as f:
                f.write(img_data)
        except:
            continue
```

```python
download_images(image_df, "Category", "Primary_Image")
```

```python
from PIL import Image

def resize_images(base_dir="images", size=(224,224)):
    for category in os.listdir(base_dir):
        cat_path = os.path.join(base_dir, category)
        for img in os.listdir(cat_path):
            try:
                path = os.path.join(cat_path, img)
                image = Image.open(path).convert("RGB")
                image = image.resize(size)
                image.save(path)
            except:
                continue
```
